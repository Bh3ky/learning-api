# Documentation

* all API documentation must always include core features. these are some of the core features:

    1. Getting Started
        - an overview or introduction that explains the purpose of the API and how consumers can interact with it. also provides the address of the API, security requirements, and instructions for requesting user IDs or API keys. 


    2. Endpoint Definitions
        - they explain the purpose of each endpoint, along with the format of requests, and any errors. 
        - an endpoint is a full address and resource name that API users will to. 


    3. OAS file
        - the OpenAPI specification (OAS) file is a machine-readable file that defines the connection instructions and endpoint defintions. 


    4. Terms of Service
        - these explain the allowed usage of the API and any rate limits or restrictions that apply. 


    5. Feedback Mechanism
        - offering a platform for people to reach and ask questions regarding the API. common platforms include support email addresses, contact forms, and GitHun issue trackers. 


    6. SDKs
        - help enforce responsible usage of the API, reducing some common problems such as overuse. 

    - other additional features include: sample program code, interactive documentation, sandbox environment, postman collection etc. 
    - Que: what is the difference between interactive API documentation and a sandbox environment??
        - interactive documentation supports test API calls, but it doesn't store the results between test calls. while a sandbox environment keeps track of previous API calls to allow us to test multiple tasks together.

    - the OAS file is automatically generated from the main.py code
